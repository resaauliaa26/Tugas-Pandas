```python
# Visualisasi
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")

df_2019 = df[df["tahun"] == 2019]

plt.figure(figsize=(14,6))
plt.bar(df_2019["nama_kabupaten_kota"], df_2019["jumlah_penderita_dm"])
plt.xticks(rotation=90)
plt.title("Jumlah Penderita DM per Kabupaten/Kota (Tahun 2019)")
plt.xlabel("Kabupaten/Kota")
plt.ylabel("Jumlah Penderita DM")
plt.tight_layout()
plt.show()

```

```python
total_tahun = df.groupby("tahun")["jumlah_penderita_dm"].sum()

plt.figure(figsize=(8,5))
plt.plot(total_tahun.index, total_tahun.values, marker="o")
plt.title("Total Penderita DM Jawa Barat per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Total Penderita DM")
plt.grid(True)
plt.show()
```

```python
top10_2019 = df_2019.sort_values("jumlah_penderita_dm", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top10_2019["nama_kabupaten_kota"], top10_2019["jumlah_penderita_dm"])
plt.gca().invert_yaxis()  # supaya ranking 1 di atas
plt.title("Top 10 Kabupaten/Kota Penderita DM Tertinggi (2019)")
plt.xlabel("Jumlah Penderita DM")
plt.ylabel("Kabupaten/Kota")
plt.tight_layout()
plt.show()
```

```python
kategori_2019 = df_2019["kategori_dm"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(kategori_2019, labels=kategori_2019.index, autopct="%1.1f%%", startangle=90)
plt.title("Proporsi Kategori DM pada Tahun 2019")
plt.show()
```
